```python
#checking null values
mushroom.isnull().sum()
```

there is no null values.

```python
#checking the values in the each feature
for i in mushroom.columns:
    print('values in',i ,':','\n',mushroom[i].value_counts())
    print('\n')
```

the 'stalk-root' column contains missing values in the form of '?'

```python
#check the relation between 'stalk-root' and target feature,'mushroom-type'
sns.countplot(mushroom['stalk-root'],hue=mushroom['mushroom-type'])
```

stalk root column has a significant amount of missing values and also it does not seem to provide useful information 
about whether a mushroom is edible or not.

```python
#dropping stalk-root column
mushroom.drop('stalk-root',axis=1,inplace=True)
```

```python
#checing the relationship of all features with target
for j in mushroom.columns:
    plt.figure(j)
    sns.countplot(mushroom[j],hue=mushroom['mushroom-type'])
```

the feature veil-type seems useless as 100% of its data contains only one type of categorical value , 'p'

```python
#dropping veil-type feature
mushroom.drop('veil-type',axis=1,inplace=True)
```

```python
mushroom.shape
```

```python
#encoding all the features
import sklearn
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for i in mushroom.columns:
    mushroom[i]=le.fit_transform(mushroom[i])
```

```python
#checking correlation of features with each other
plt.figure(figsize=(20,10))
sns.heatmap(mushroom.corr(),annot=True)
```

```python
#selecting all weakly correlated features
weak_features=[]
for i in mushroom.columns:
    corr=abs(mushroom['mushroom-type'].corr(mushroom[i]))
    if corr<0.1:
        weak_features.append(i)
print(weak_features)
```

```python
#dropping weak features
mush=mushroom.drop(weak_features,axis=1)
mush.shape
```

```python
#seperating target variable from the data
x=mush.drop('mushroom-type',axis=1)
y=mush['mushroom-type']
x.shape,y.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier,AdaBoostClassifier,GradientBoostingClassifier
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
from sklearn.model_selection import KFold,cross_val_score,GridSearchCV,train_test_split
```

```python
#train test split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=4)
```

```python
#choosing the best algorithm
#train and predict the data using cross-validation
model_list=[LogisticRegression(),KNeighborsClassifier(),GaussianNB(),SVC(),DecisionTreeClassifier(),RandomForestClassifier(),
      AdaBoostClassifier(),GradientBoostingClassifier()]
for model in model_list:
    kfold=KFold(n_splits=10,random_state=None)
    results=cross_val_score(model,x_train,y_train,cv=kfold,scoring='accuracy')
    print(model)
    print(results.mean())
    print('\n')
```

clearly RandomForest ,GredientBoosting and DecisionTree worked equally best.

```python
#predicting the untrained data to get the actual accuracy of these best models
best_models=[DecisionTreeClassifier(),RandomForestClassifier(),GradientBoostingClassifier()]
for m in best_models:
    m.fit(x_train,y_train)
    pred=m.predict(x_test)
    print(m)
    print(accuracy_score(pred,y_test))
    print('\n')
```

all these models are working best.

```python
#selecting RandomForest for final training
rfc=RandomForestClassifier()
rfc.fit(x,y)
```

```python
#checking feature importances
imp=pd.Series(rfc.feature_importances_,index=x.columns)
imp.plot(kind='bar')
plt.show()
```

```python
#saving the model using joblib
import joblib 
joblib.dump(rfc,'project_5_mushroom.pkl')
```
