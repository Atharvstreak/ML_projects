# DATA ANALYSIS OF CRIME AGAINST WOMEN IN INDIA IN 2013

# Import needed modules and get the required dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
#getting the crime against women in 2013 dataset
crime_data=pd.read_csv('Downloads/cprsvCAW_2_2013.csv')
crime_data
```

```python
#checking the total categories in crime
crime_data['CRIME'].value_counts()
```

There are total 11 different types of crimes against women in India

```python
#remove Total UTs,States,All-India rows
temp=crime_data.groupby(['STATE/UT','CRIME']).sum()
temp.drop(index=['Total (All-India)','Total (States)','Total (UTs)'],inplace=True)
temp.head(12)
```

```python
#groupby crime and then group by states
grouped_crime=temp.groupby(['CRIME','STATE/UT']).sum()
grouped_crime
```

```python
#create a df which only contains top 10 states with highest total people arrested for crime against women
total_crime=pd.DataFrame(grouped_crime.loc['Total Crimes against Women',:])
top_10=total_crime.sort_values(['PERSONS ARRESTED'],axis=0,ascending=False).head(10)
top_10
```

These are the top 10 states in India that have high numbers of crimes against women, than other parts of India. 

```python
#visualising the crime of 'Assault on women with intent to outrage her modest' in each state
grouped_crime.loc['Assault on women with intent to outrage her Modest',:].plot(kind='bar',stacked=True,figsize= (10, 5))
plt.style.use('seaborn-whitegrid')
plt.title('Crime : Assault on women with intent to outrage her Modest')
plt.xticks(rotation=90)
plt.show()
```

The crime of Assault on women with intent to outrage her Modest :                                                            

1)The state having highest number of registered cases is Madhya Pradesh and Maharashtra, Uttar Pradesh had almost the same number of registered cases as MP and Maharashtra.The places with least number of crimes in this category are not states but all the islands.

2)The total number of people arrested in this crime is greater than the total registered cases. This can be seen in almost all places.  

3)The total people convicted are very less among the total people arrested ,it seems like most of the arrested people were chargesheeted instead.

4)Delhi has less population than many states like Tamil Nadu,still has relatively higher crime in this category. 

5)The number of crimes in this category is more in the Northern part of India as compared to the Southern part.
