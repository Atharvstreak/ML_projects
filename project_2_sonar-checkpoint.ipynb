```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
#get the data
sonar_data = pd.read_csv('Downloads/sonar.csv',header=None)
sonar_data.head()
```

```python
#explore the data
sonar_data.info(),sonar_data.isnull().sum()
```

all columns are of float64 data type except last column, which is object data type and there are no NaN values.

```python
#statistical description
sonar_data.describe()
```

```python
#counting total number of rocks and mines from the last column
sonar_data.iloc[:,-1].value_counts()
```

total mines = 111 and rocks = 97

```python
#checking the skewness coefficients of the data
sonar_data.skew()
```

```python
#check for zeroes in all the data 
(sonar_data.values==0).any()
```

```python
#replace all zeroes with 0.0000001
sonar_data = sonar_data.mask(sonar_data==0,0.0000001)
```

```python
(sonar_data.values==0).any()
```

```python
#normally distributing the values of all features 
from scipy.stats import boxcox
for i in sonar_data.columns:
    sonar_data[i]=boxcox(sonar_data[i])[0]
    if i==59:
        break
sonar_data.skew() 
```

```python
#visualizing the normal distribution of values in each column
plt.figure(figsize=(6,3))
for i in sonar_data.columns:
    plt.figure(i)
    sns.distplot(sonar_data[i])
    if i==59:
        break
```

```python
#standardizing all the features to bring their values in a common scale
import sklearn
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
scaled_sonar_data = pd.DataFrame(scale.fit_transform(sonar_data.iloc[:,0:-1]),columns=sonar_data.iloc[:,0:-1].columns)
new_sonar_data = pd.concat([scaled_sonar_data,sonar_data.iloc[:,-1]],axis=1)
new_sonar_data.head()
```

```python
#encoding last column
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
new_sonar_data.iloc[:,-1]=le.fit_transform(new_sonar_data.iloc[:,-1])
```

```python
new_sonar_data.head()
```

```python
#detecting the outliers
import scipy
from scipy.stats import zscore
z = abs(zscore(new_sonar_data))
z[z>3]
```

```python
#removing the outliers
sonar = new_sonar_data[(z<3).all(axis=1)]
```

```python
sonar.head() #data without the outliers
```

```python
sonar_data.shape,sonar.shape #change in shape after removing outliers
```

```python
#exploring correlation
plt.figure(figsize=(12,6))
sns.heatmap(sonar.corr())
```

```python
#split target and input variables
x = sonar.iloc[:,0:-1] #input variables
y = sonar.iloc[:,-1] #target variable
```

```python
x.shape,y.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier,AdaBoostClassifier,GradientBoostingClassifier
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
from sklearn.model_selection import KFold,cross_val_score,GridSearchCV
```

```python
#choosing the best algorithm
#train and predict the data using cross-validation
model_list=[LogisticRegression(),KNeighborsClassifier(),GaussianNB(),SVC(),DecisionTreeClassifier(),RandomForestClassifier(),
      AdaBoostClassifier(),GradientBoostingClassifier()]
for model in model_list:
    kfold=KFold(n_splits=10,random_state=None)
    results=cross_val_score(model,x,y,cv=kfold,scoring='accuracy')
    print(model)
    print(results.mean())
    print('\n')
```

Logistic and AdaBoost are working equally, so lets apply both and select the best out of them

```python
#applying logisticRegression 
lg = LogisticRegression()
lg.fit(x,y)
lg.score(x,y)
```

```python
#Applying AdaBoostClassifier
ada = AdaBoostClassifier()
ada.fit(x,y)
ada.score(x,y)
```

To avoid overfitting , we use grid_search_cv for hyperparameter tuning

```python
params = {'n_estimators':range(50,500,50),
          'algorithm':['SAMME.R','SAMME']}
ada_cv=GridSearchCV(estimator=ada,param_grid=params,cv=5)
ada_cv.fit(x,y)
```

```python
print(ada_cv.best_params_)
```

```python
#using best params
ada_new = AdaBoostClassifier(algorithm='SAMME.R',n_estimators=200)
ada_new.fit(x,y)
```

```python
ada_new.score(x,y)
```

```python
pred=ada_new.predict(x)
```

```python
print(accuracy_score(pred,y))
```

```python
print(confusion_matrix(pred,y))
```

```python
print(classification_report(pred,y))
```

it seems that AdaBoostClassifier has learned all the data very well!
